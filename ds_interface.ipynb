# 1학기

```python
from selenium.webdriver.common.keys import Keys

keyword = '[DS Interface]'
url1 = 'https://www.youtube.com/playlist?list=PLVUMrRRWzO2pSbuW-RhxroVl0xhgycZvb'.format(keyword)

driver1 = wd.Chrome(r"C:\Users\chromedriver.exe")
driver1.get(url1)
```

```python
soup1 = BeautifulSoup(driver1.page_source, 'lxml')
title1 =soup1.select('a#video-title')
# 영상 제목만 전체 조회
for i in title1:
    i.text.strip()
```

```python
title_list1 = []

for i in range(len(title1)):
    title_list1.append(title1[i].text.strip().replace("[DS Interface]", ""))
    
print(title_list1)
```

```python
info1 = {'2021학년도 1학기':title_list1}
pd.DataFrame(info1)
```
