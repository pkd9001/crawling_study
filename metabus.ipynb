```python
import pandas as pd
import numpy as np
import time
import re
from selenium import webdriver as wd
from bs4 import BeautifulSoup
from datetime import datetime
from tqdm import tqdm
import requests
```

```python
keyword = input('입력하세요 : ')
```

```python
driver = wd.Chrome('C:/Users/user/chromedriver_win32/chromedriver') # + reward
driver.implicitly_wait(1)
url = f'https://patents.google.com/?q={keyword}&oq={keyword}'
driver.get(url)

item = []
result = []

def get_inventor(soup):
    string = ""
    try:
        inventors = soup.select('dd[itemprop=inventor]')
        for inventor in inventors:
            if inventor.text.strip() not in string:
                string += inventor.text.strip() + ','
        string = string[:-1]
    except:
        print('Inventor이 존재하지 않습니다.')
        return None
    return string

for n in range(3):
    for i in range(10):
        driver.find_element_by_xpath(f'html/body/search-app/search-results/search-ui/div/ \
                                    div/div[2]/div/div/div[1]/section/search-result-item[{i+1}]/ \
                                    article/state-modifier/a/h3/raw-html/span').click()

        title_path = '/html/body/search-app/search-result/search-ui/div/div/div/div/div/ \
                      result-container/patent-result/div/div/div/h1'
        num_path = '/html/body/search-app/search-result/search-ui/div/div/div/div/div/ \
                    result-container/patent-result/div/div/div/div[1]/div[2]/section/header/h2'
        country_path = '/html/body/search-app/search-result/search-ui/div/div/div/div/div/ \
                        result-container/patent-result/div/div/div/div[1]/div[2]/section/header/p'
        inventor_path = driver.current_url

        title = driver.find_element_by_xpath(title_path)
        title_result = title.text
        num = driver.find_element_by_xpath(num_path)
        num_result = num.text
        country = driver.find_element_by_xpath(country_path)
        country_result = country.text

        response = requests.get(inventor_path)
        soup = BeautifulSoup(response.content, 'html.parser')

        inventor_result = get_inventor(soup)

        item.extend([title_result,num_result,country_result,inventor_result])
        result.append(item)
        item = []
        driver.back()
    driver.find_element_by_xpath('/html/body/search-app/search-results/ \
                                search-ui/div/div/div[2]/div/div/div[1]/div[5]/search-paging/ \
                                state-modifier[3]/a/paper-icon-button/iron-icon').click()
driver.close()
data = pd.DataFrame(result)
data.columns = ["title","patent number","country","inventor"]
```
